```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data = sns.load_dataset('titanic')
print(titanic_data.head())
```

```python
survivors_by_sex = titanic_data[titanic_data['survived'] == 1]['sex'].value_counts()

sns.barplot(
    x = survivors_by_sex.index, 
    y = survivors_by_sex.values, 
    color='grey',
    edgecolor='black'
)
plt.title('Number of Survivors by Sex')
plt.xlabel('Sex')
plt.ylabel('Count of Survivors')
plt.show()
```

```python
survivors_by_class = titanic_data[titanic_data['survived'] == 1]['pclass'].value_counts()

sns.barplot(
    x = survivors_by_class.index, 
    y = survivors_by_class.values, 
    color='lightgreen',
    alpha=0.7              
)
plt.title('Number of Survivors by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count of Survivors')
plt.show()
```

```python
survivors_by_deck = titanic_data[titanic_data['survived'] == 1]['deck'].value_counts()

sns.barplot(x = survivors_by_deck.index, y = survivors_by_deck.values)
plt.title('Number of Survivors by Deck')
plt.xlabel('Deck')
plt.ylabel('Count of Survivors')
plt.show()
```
